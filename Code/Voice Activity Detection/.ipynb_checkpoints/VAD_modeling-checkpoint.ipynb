# Train a Voice Activity Detector using Deep Nural Networks
Here we will be using the Mel spectrogram features generated by the file 'VAD_dataset_generator.ipynb' to train a Convolutional Nural Network, to distinguish the speech and non-speech segments in the audio.

## Loading Libraries and Training Data

```python
%pylab inline
import io
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from tensorflow import keras
from tensorflow.keras import layers
from matplotlib import pyplot as plt
from keras.utils.vis_utils import plot_model
from sklearn.manifold import TSNE
```

```python
X = np.load('./vad_dataset/vad_mel_collection_train.npy')       #The training data are the mel spectrogram features of 1 second segments of the audio files
y = np.load('./vad_dataset/vad_labels_train.npy')               #Labels represent the speaker number
X=X.reshape(-1,44,128,1)
print(X.shape,y.shape)
```

```python
# Code for equal partitions of both type of labels datasets

a=0
b=0
for i in range(y.shape[0]):
  if y[i]==1:
    a+=1
  else:
    b+=1
c = min(a,b)

X_balanced = []
y_balanced = []

done0,done1 = 0,0
for i in range(y.shape[0]):
  if y[i]==1 and done1<c:
    X_balanced.append(X[i])
    y_balanced.append(y[i])
    done1+=1

  if y[i]==0 and done0<c:
    X_balanced.append(X[i])
    y_balanced.append(y[i])
    done0+=1

X_balanced = np.array(X_balanced)
y_balanced = np.array(y_balanced)

print(y_balanced.shape)
```

##Creating and Training CNN model

```python
X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.1, shuffle=True)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(filters=8, kernel_size=5, padding='same', activation='relu',input_shape = (44,128,1)),
    tf.keras.layers.MaxPooling2D(pool_size=2),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Conv2D(filters=16, kernel_size=5, padding='same', activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=2),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Conv2D(filters=16, kernel_size=5, padding='same', activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=2),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Conv2D(filters=32, kernel_size=5, padding='same', activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=2),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(1, activation='sigmoid'), # Sigmoid activation on final dense layer for binary classification
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate = 0.0005),       # Using Adam optimizer
    loss = tf.keras.losses.binary_crossentropy,
      metrics=['accuracy']
)
```

```python
history = model.fit(X_train,y_train,epochs=30, validation_data=(X_test,y_test))
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('number of epoch')
plt.legend(['train', 'val'], loc='upper right')
plt.show()
```

## Testing the VAD trained

```python
def testing(filenames,con_matrix = False):
  avg_acc = 0.0
  for filename in filenames:
    X_testing = np.load('./vad_dataset/vad_mel_collection_test'+filename+'.npy')
    y_testing = np.load('./vad_dataset/vad_labels_test'+filename+'.npy')
    X_testing = X_testing.reshape(-1,44,128,1)
    result = model.evaluate(X_testing,y_testing)    # Getting predicted values from the model, output 1 and 0 represent speech and non-speech respectively
    avg_acc = avg_acc + result[1]

    if con_matrix:
      y_pred = model.predict(X_testing)
      y_pred = np.where(y_pred > 0.5, 1, 0)
      matrix = confusion_matrix(y_testing, y_pred)
      print('\nConfusion Matrix\n',matrix,'\n')

  avg_acc = avg_acc/len(filenames)
  print("\nAverage Accuracy for unseen audio files: "+str(avg_acc))      # Output the average accuracy over all the testing audio files
```

```python
filenames = ['IB4011','IS1000c','TS3010c']
testing(filenames,con_matrix = True)
```
